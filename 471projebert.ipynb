```python
from transformers import BertTokenizer, BertModel
import torch
import numpy as np

# Load pre-trained model and tokenizer
model_name = 'bert-base-uncased'
#model_name = 'bert-base-chinese'
#model_name = 'dbmdz/bert-base-turkish-uncased'

tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertModel.from_pretrained(model_name, output_hidden_states=True)

# Example sentence and POS tags
sentence = "The quick brown fox jumps over the lazy dog"
pos_tags = ["DET", "ADJ", "ADJ", "NOUN", "VERB", "ADP", "DET", "ADJ", "NOUN"]

# Convert POS tags to numerical labels (this should be consistent across your dataset)
pos_tag_to_id = {"DET": 0, "ADJ": 1, "NOUN": 2, "VERB": 3, "ADP": 4}
labels = [pos_tag_to_id[tag] for tag in pos_tags]



labels = [pos_tag_to_id[tag] for tag in pos_tags]

# Tokenize the sentence
inputs = tokenizer(sentence, return_tensors='pt')

# Forward pass to get hidden states
with torch.no_grad():
    outputs = model(**inputs)
    hidden_states = outputs.hidden_states  # tuple of (num_layers, batch_size, seq_length, hidden_size)

# Choose a specific layer (e.g., layer 3)
layer_idx = 3
layer_output = hidden_states[layer_idx]  # (batch_size, seq_length, hidden_size)

# Extract features for the tokens (excluding special tokens like [CLS] and [SEP])
token_features = layer_output[0, 1:-1, :]  # (seq_length - 2, hidden_size)

# Convert token features to numpy array
token_features_np = token_features.numpy()
```

```python
from sklearn.linear_model import LogisticRegression

# Define the probing model
probe = LogisticRegression(max_iter=1000)

# Train the probing model
probe.fit(token_features_np, labels)

# Make predictions on the same data
predictions = probe.predict(token_features_np)

# Print predictions
print("Predicted labels:", predictions)

# Convert predictions back to POS tags
id_to_pos_tag = {v: k for k, v in pos_tag_to_id.items()}
predicted_tags = [id_to_pos_tag[pred] for pred in predictions]

print("Predicted POS tags:", predicted_tags)
```

```python
from sklearn.metrics import accuracy_score

# Calculate accuracy
accuracy = accuracy_score(labels, predictions)
print(f"Accuracy: {accuracy * 100:.2f}%")
```

```python
import stanza
from graphviz import Digraph

# Download and load the Chinese model
stanza.download('zh')
nlp = stanza.Pipeline(lang='zh', processors='tokenize,pos,lemma,depparse')

text_ch = "我喜欢学习自然语言处理"
doc_ch = nlp(text_ch)

# Function to visualize the dependency tree
def visualize_dependency_tree(doc):
    dot = Digraph()

    for sentence in doc.sentences:
        for word in sentence.words:
            dot.node(str(word.id), word.text)
            if word.head > 0:
                dot.edge(str(word.head), str(word.id), label=word.deprel)

    return dot

# Extract and print dependency relations
for sentence in doc_ch.sentences:
    print(f"Sentence: {sentence.text}")
    for word in sentence.words:
        print(f"Word: {word.text}, Head: {word.head}, Relation: {word.deprel}")

dot = visualize_dependency_tree(doc_ch)
dot.render('dependency_tree_ch', format='png', view=True)  
```
