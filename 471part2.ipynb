```python
def pos_tag_sentences(nlp, sentences):
    tagged_sentences = []
    for sentence in sentences:
        doc = nlp(" ".join([token for token, _ in sentence]))
        tagged_sentence = [(token.text, token.pos_) for token in doc]
        tagged_sentences.append(tagged_sentence)
    return tagged_sentences
```

```python
from graphviz import Digraph

def visualize_dependency_tree(doc):
    dot = Digraph()

    for sentence in doc.sentences:
        for word in sentence.words:
            dot.node(str(word.id), word.text)
            if word.head > 0:
                dot.edge(str(word.head), str(word.id), label=word.deprel)

    return dot

# dependency relations
for sentence in doc_ch.sentences:
    print(f"Sentence: {sentence.text}")
    for word in sentence.words:
        print(f"Word: {word.text}, Head: {word.head}, Relation: {word.deprel}")

# dependency tree
dot = visualize_dependency_tree(doc_ch)
dot.render('dependency_tree', format='png', view=True)  # Saves and opens the file
```

dependency trees witih udpipe

```python
import ufal.udpipe
```

```python


# Load the UDPipe model
def load_model(model_path):
    model = ufal.udpipe.Model.load(model_path)
    if not model:
        raise Exception("Cannot load model from path: " + model_path)
    return model

# Tokenize, tag, parse, and lemmatize the text
def process_text(model, text):
    pipeline = ufal.udpipe.Pipeline(model, 'tokenize', 'tag', 'parse', 'conllu')
    processed = pipeline.process(text)
    return processed

# Print dependency trees
def print_dependency_trees(processed):
    for line in processed.split('\n'):
        if line and not line.startswith('#'):
            cols = line.split('\t')
            if len(cols) == 10:
                print(f"{cols[1]}\t{cols[7]}\t{cols[6]}")

# Main function
def main():
    model_path = 'english-ud-2.5-191206.udpipe'  # Path to the UDPipe model
    model = load_model(model_path)
    
    text_english = "The quick brown fox jumps over the lazy dog."
    processed = process_text(model, text_english)
    
    print("Dependency trees:")
    print_dependency_trees(processed)

if __name__ == "__main__":
    main()
```

```python
from transformers import BertTokenizer, BertModel

# Load pre-trained BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained(model_name, output_attentions=True)
```

```python
text_sample = "I think Andre said the Jamaica Observer reported that Usain Bolt broke the 100m record."
inputs = tokenizer(text_sample, return_tensors='pt')
inputs
```

```python
with torch.no_grad():
    outputs = model(**inputs)

# The hidden states are in the outputs[0]
# outputs[0] is a tensor of shape (batch_size, sequence_length, hidden_size)
hidden_states = outputs.last_hidden_state
hidden_states
```

```python
# Get the embeddings of the [CLS] token (representing the whole sentence)
cls_embedding = hidden_states[0][0]  # Assuming batch size 1
len(cls_embedding)
```

```python
outputs2 = model(**inputs)

# Extract attention weights
attention = outputs2.attentions  # List of tensors with shape (batch_size, num_heads, seq_length, seq_length)
from bertviz import head_view
head_view(attention, tokenizer.convert_ids_to_tokens(inputs['input_ids'][0]))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def get_sentence_embedding(sentence):
    inputs = tokenizer(sentence, return_tensors='pt')
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze().numpy()

sentence1 = "I love programming."
sentence2 = "Wikipedia is a free online encyclopedia that anyone can edit, and millions already have."

embedding1 = get_sentence_embedding(sentence1)
embedding2 = get_sentence_embedding(sentence2)

similarity = cosine_similarity([embedding1], [embedding2])
print("Similarity:", similarity[0][0])
```
